1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
import numpy as np

salary = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

# среднее арифметическое
mean_salary = sum(salary)/len(salary)
print(f'Среднее арифметическое: {mean_salary}')

# среднее квадратичное отклонение

sum_mean = 0
for element in salary:
    sum_mean = (element - mean_salary)**2
    sum_mean += sum_mean 
    

std_salary = np.sqrt(sum_mean/len(salary))
print(f'Среднее квадратичное отклонение {std_salary}')

# смещенная дисперсия
variance_salary = sum_mean/len(salary)
print(f'Смещенная дисперсия {variance_salary}')

# несмещенная дисперсия
variance_salary2 = sum_mean/(len(salary) - 1)
print(f'Несмещенная дисперсия {variance_salary2}')
```

2.В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
from math import factorial

# функция расчета числа сочетаний
def combinations(n, k):
    return int(factorial(n)/(factorial(k) * factorial(n - k)))

P1 = (combinations(5, 2)/combinations(8, 2)) * ((combinations(5,1)*combinations(7, 3))/combinations(12, 4))
print(f'P1 {P1}')

P2 = ((combinations(5, 1) * combinations(3, 1))/combinations(8, 2))*((combinations(5, 2)*combinations(7,2))/combinations(12, 4))
print(f'P2 {P2}')

P3 = (combinations(3,2) / combinations(8,2)) * ((combinations(5, 3) * combinations(7, 1))/combinations(12, 4))
print(f'P3 {P3}')

P = P1 + P2 + P3

print (f'Вероятность, что 3 мяча белые {P}')
```

3.На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
p1 = 0.9
p2 = 0.8
p3 = 0.6
n = 3
k = 1

# Полная вероятность события B
P_b = (k/n) * p1 + (k/n) * p2 + (k/n) * p3
#print(P_b)

# Априорная вероятность
P_a = k/n

# Вероятность, что выстрел произведен первым спортсменом
P_ab1 = (p1 * P_a)/P_b
print(f'Вероятность, что выстрел произведен первым спортсменом {P_ab1}')

# Вероятность, что выстрел произведен вторым спортсменом
P_ab2 = (p2 * P_a)/P_b
print(f'Вероятность, что выстрел произведен первым спортсменом {P_ab2}')

# Вероятность, что выстрел произведен третим спортсменом
P_ab3 = (p3 * P_a)/P_b
print(f'Вероятность, что выстрел произведен первым спортсменом {P_ab3}')
```

5,Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
P1 = 0.1
P2 = 0.2
P3 = 0.25

#a) Вероятность, что выйдут из строя все детали
P = P1 * P2 * P3
print(f'Вероятность, что выйдут из строя все детали {P}')

#b Вероятность, что выйдут из строя только две детали
P = P1 * P2 * (1- P3) + P1 * P3 * (1 - P2) + P2 * P3 * (1 - P1)
print(f'Вероятность, что выйдут из строя только две детали {P}')

#c Вероятность, что выйдет из строя хотя бы одна деталь
P = 1 - (1 - P1)*(1 - P2)*(1 - P3)
print(f'Вероятность, что выйдет из строя хотя бы одна деталь {P}')
```
